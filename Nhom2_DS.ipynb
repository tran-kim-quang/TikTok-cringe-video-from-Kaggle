```python
try:
    driver.find_element(By.XPATH,'//*[@id="site-container"]/div/div[4]/div[2]/div[1]/div/div/input').click()
    driver.find_element(By.XPATH,'//*[@id="site-container"]/div/div[5]/div[1]/div/div[1]/form/input').send_keys('tiktok')
    driver.find_element(By.XPATH,'//*[@id="site-container"]/div/div[5]/div[1]/div/div[2]/button[1]').click()
    search=driver.find_elements(By.XPATH,'//*[@id="site-container"]/div/div[5]/div[2]/div/div/div[2]/div[2]/ul/a')
except NoSuchElementException:
       print('Element not found')
```

```python
def crawl_data(url):
    driver.get(url)
    try:
        driver.find_element(By.XPATH,'//*[@id="site-content"]/div[2]/div/div[2]/div[1]/div/a/button/span[1]').click()
    except NoSuchElementException:
       print('Element not found')
```

```python
for i in lst_data:
    crawl_data(i)
```

### Trực quan hoá dữ liệu

#### Đánh giá độ tích cực của các bài đăng theo 'score'

```python
def rate(x: float):
    return 'Negative' if x < -0.5 else 'Positive' if x > 0.5 else 'Neutral'
data['rate']=[rate(x) for x in data['score']]
labels=data['rate'].value_counts().keys()
y=data['rate'].value_counts().values
plt.pie(y,labels=labels,autopct='%1.1f%%')
plt.show()
```

* **Chỉ số tích cực(có giá trị mục đích) của video trên tiktok chiếm phần lớn các video, về mức trung bình thì lại ít hơn mức tiêu cực**
> Vậy các bài đăng trên tiktok có 2 hình thái chủ yếu là đem lại lợi ích và không có mục đích gì

#### Đánh giá theo 'tag'

```python
rate_tag=data['tag'].value_counts()
rate_tag
```

```python
y=rate_tag.values
labels=rate_tag.keys()
plt.bar(labels,y)
plt.xticks(rotation=45)
plt.show()
```

* **Nhiều nhất vẫn là các bài mang tính 'cringe' theo tính tiêu cực, còn lại là các bài đăng mang tính về thảo luận hoặc hài hước và 1 chút đùa về chính trị**

**So sánh hai biểu đồ đánh giá giữa 'score' và 'tag' thì lại thấy 2 phần chiếm nhiều nhất là bài đăng mang tính 'cringe' và tương đương nó là mục tích cực**
